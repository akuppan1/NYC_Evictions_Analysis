```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Evictions.csv')
```

```python
data
```

```python
data.info()
```

```python
data['YEARLY DATE'] = 0
```

```python
for i in range(len(data['Executed Date'])):
    data['YEARLY DATE'][i] = data['Executed Date'][i][6:10] 
```

```python
data['YEARLY DATE'].value_counts()
```

```python
# data['YEARLY DATE'].plot(kind= 'bar', x=['2017','2018','2019','2020','2021'])
```

```python
data_2017 = data[data['YEARLY DATE'] == 2017]
data_2018 = data[data['YEARLY DATE'] == 2018]
data_2019 = data[data['YEARLY DATE'] == 2019]
data_2020 = data[data['YEARLY DATE'] == 2020]
data_2021 = data[data['YEARLY DATE'] == 2021]
```

```python
print('2017 per borough')
print(data_2017['BOROUGH'].value_counts())
print('2018 per borough')
print(data_2018['BOROUGH'].value_counts())
print('2019 per borough')
print(data_2019['BOROUGH'].value_counts())
print('2020 per borough')
print(data_2020['BOROUGH'].value_counts())
print('2021 per borough')
print(data_2021['BOROUGH'].value_counts())
```

```python

fig, axes = plt.subplots(nrows=2, ncols=2)
fig.tight_layout(h_pad=8)
# df1.plot(ax=axes[0,0])
data_2021['BOROUGH'].value_counts().plot(ax = axes[0,0],kind ='bar', title='2021 Evictions Per Borough')
data_2020['BOROUGH'].value_counts().plot(ax = axes[0,1],kind ='bar', title='2020 Evictions Per Borough')
data_2019['BOROUGH'].value_counts().plot(ax = axes[1,0],kind ='bar', title='2019 Evictions Per Borough')
data_2018['BOROUGH'].value_counts().plot(ax = axes[1,1],kind ='bar', title='2018 Evictions Per Borough')
```
